# relation 합쳐서 하나의 dataframe으로 합치기

```python
#relationship dataframe에 연관된 직원끼리 쭉 이어 붙이기
relationship = pd.DataFrame(columns = ['직원1_행번','직원2_행번'])
relationship
```

```python
new_relation1=[]
for i in relation1:
    if len(i)>=2:
        new_relation1.append(i)
new_relation1 = pd.Series(new_relation1)

new_relation2=[]
for i in relation2:
    if len(i)>=2:
        new_relation2.append(i)
new_relation2 = pd.Series(new_relation2)

```

```python
new_relation3=[]
for i in relation3:
    if len(i)>=2:
        new_relation3.append(i)
new_relation3 = pd.Series(new_relation3)
```

```python
relation_tot = pd.concat([new_relation1, new_relation2, new_relation3])
```

```python
relation_tot.head()
```

```python
for rel in relation_tot:
    temp_rel = sorted(rel)
    relationship = relationship.append(pd.DataFrame(list(itertools.combinations(temp_rel,2)),columns = relationship.columns))
```

```python
relationship.head()
```

```python
relationship = relationship.groupby(relationship.columns.tolist(),as_index=False).size()
```

```python
relationship.columns = ['직원1', '직원2', 'count']
```

```python
relationship["count"].unique().max()
```

```python
relationship
```

```python
relationship['count']= (relationship['count'] - relationship['count'].min())/(relationship['count'].max()-relationship['count'].min())
```
